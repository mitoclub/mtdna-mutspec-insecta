```python
import pandas as pd
from sklearn import preprocessing
```

```python
leptraits = pd.read_csv('/home/gabs/Documents/lab/TermitesAndCockroaches/mtdna-mutspec-insecta/data/EcoMorphDBs/leptraits.csv', low_memory=False)
meta = pd.read_csv('/home/gabs/Documents/lab/TermitesAndCockroaches/mtdna-mutspec-insecta/data/NemuPipeline/AllInsects/midori_all_insects_mutspec/msMetaData.tsv', sep='\t')
glad = pd.read_csv('/home/gabs/Documents/lab/TermitesAndCockroaches/mtdna-mutspec-insecta/data/EcoMorphDBs/global_ants.csv', encoding='latin1')
animal_traits = pd.read_csv('/home/gabs/Documents/lab/TermitesAndCockroaches/mtdna-mutspec-insecta/data/EcoMorphDBs/animal_traits.csv')
diptera_traits = pd.read_csv('/home/gabs/Documents/lab/TermitesAndCockroaches/mtdna-mutspec-insecta/data/EcoMorphDBs/DipteraTraitDataset.csv')
cock_terms_skew = pd.read_csv('/home/gabs/Documents/lab/TermitesAndCockroaches/mtdna-mutspec-insecta/data/DescriptiveStat/midori_Blattodea_skew.csv', index_col=0)
diptera_skew = pd.read_csv('/home/gabs/Documents/lab/TermitesAndCockroaches/mtdna-mutspec-insecta/data/DescriptiveStat/midori_Diptera_skew.csv', index_col=0)
```

```python
meta.Species = meta.Species.apply(lambda sp: sp.split('_'))
meta.Species = meta.Species.apply(lambda sp: " ".join(sp))
```

```python
meta.head()
```

```python
leptraits['Species'] = leptraits['Genus'] + ' ' + leptraits['Species']
leptraits = leptraits.drop(columns='Genus')
```

```python
df = []
for sp in meta['Species']:
    if sp in leptraits['Species'].to_list():
        leptraits_data = leptraits.loc[leptraits['Species'] == sp]
        df.append(leptraits_data.assign(Family=meta.loc[meta['Species'] == sp, 'Family'].to_list() * len(leptraits_data)))
df = pd.concat(df)
```

```python
df = []
for sp in meta['Species']:
    if sp in diptera_traits['scientificName'].to_list():
        diptera_traits_data = diptera_traits.loc[diptera_traits['scientificName'] == sp]
        df.append(diptera_traits_data.assign(Family=meta.loc[meta['Species'] == sp, 'Family'].to_list() * len(diptera_traits_data)))
df = pd.concat(df)
```

```python
t = df [['scientificName',	'habitat',	'traitName',	'traitValue']]
```

```python
print(t['habitat'].unique())
t['traitValue'].unique()
```

```python
label_encoder = preprocessing.LabelEncoder()
t['habitat']=label_encoder.fit_transform(t['habitat'])
t['traitValue']=label_encoder.fit_transform(t['traitValue'])
```

```python
print(t['habitat'].unique())
t['traitValue'].unique()
```

```python
t = t.rename(columns={'scientificName' : 'species'})
t.to_csv('/home/gabs/Documents/lab/TermitesAndCockroaches/mtdna-mutspec-insecta/data/EcoMorphDBs/DipteraTrait_label_encoded.csv', index=False)
```
